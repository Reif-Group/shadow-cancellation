```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import matplotlib
import os
```

```python
def read_lines(file):
    with open(file, 'r') as fin:
        lines = fin.readlines()
    lines = [line.strip() for line in lines]
    return lines
```

```python
def plot_combined(files, special_labels=[], plot_index=1, styles=[]):
    
    lines_list = [read_lines(file) for file in files]
    labels = [lines[0].split() for lines in lines_list]
    for i in range(len(special_labels)):
        labels[i] = ['time'] + [l+'_'+ special_labels[i] for l in labels[i][1:]]
    datas = np.array([[line.split() for line in lines[1:]] for lines in lines_list])
    dfs = [pd.DataFrame(data=datas[i], columns=labels[i], dtype=np.float32) for i in range(len(datas))]
    ax = plt.axes()
    for i in range(len(dfs)):
        dfs[i].plot(x='time', y=labels[i][plot_index], ax=ax, ylabel='Leak concentration')
```

```python
plot_combined(['rps/0_0/plot_o_s_p_nc', 'rps/0_2/plot_o_s_p_nc', 'rps/0_9/plot_o_s_p_nc'], special_labels=['0.0', '0.2', '0.9'], plot_index=1)
```

```python
plot_combined(['uniamp/leaky/plot_leaky_2e-2', 'uniamp/leaky/plot_leaky_2e-5', 
               'uniamp/leaky/plot_leaky_2e-6', 'uniamp/leaky/plot_leaky_2e-7'],
              special_labels=['2e-2', '2e-5', '2e-6', '2e-7'], plot_index=2)
```

```python
import os 
import shutil

main_folders = ['uniamp', 'biamp', 'rps', 'con']
subfolders  = ['original', 'leaky', 'leaky_shadow']
HOME_DIR = './'
```

```python
if not os.path.exists(os.path.join(HOME_DIR, 'all_results')):
    os.mkdir(os.path.join(HOME_DIR, 'all_results'))

OUT_DIR = os.path.join(HOME_DIR, 'all_results')
```

```python
for main_f in main_folders:
    for sub_f in subfolders:
        imgname = '_'.join([main_f, sub_f]) + '.png'
        src = os.path.join(HOME_DIR, main_f, sub_f, "images", 'main.png')
        dest = os.path.join(OUT_DIR, imgname)
        shutil.copy(src, dest)
```
