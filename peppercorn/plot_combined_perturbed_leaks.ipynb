```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import matplotlib
```

```python
def read_lines(file):
    with open(file, 'r') as fin:
        lines = fin.readlines()
    lines = [line.strip() for line in lines]
    return lines
```

```python
def plot_combined(files, special_labels=[], plot_index=1):
    
    lines_list = [read_lines(file) for file in files]
    labels = [lines[0].split() for lines in lines_list]
    for i in range(len(special_labels)):
        labels[i] = ['time'] + [l+'_'+ special_labels[i] for l in labels[i][1:]]
    datas = np.array([[line.split() for line in lines[1:]] for lines in lines_list])
    dfs = [pd.DataFrame(data=datas[i], columns=labels[i], dtype=np.float32) for i in range(len(datas))]
    ax = plt.axes()
    for i in range(len(dfs)):
        dfs[i].plot(x='time', y=labels[i][plot_index], ax=ax, ylabel='Leak concentration')
```

```python
plot_combined(['rps/0_0/plot_o_s_p_nc', 'rps/0_2/plot_o_s_p_nc', 'rps/0_9/plot_o_s_p_nc'], special_labels=['0.0', '0.2', '0.9'], plot_index=1)
```
